3. –ü—Ä–æ–∏–∑–≤–µ–¥–∏—Ç–µ **–ø–∞—Ä—Å–∏–Ω–≥** –≤—ã–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

- –∏–∑–≤–ª–µ–∫–∏—Ç–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ–¥ –ø–æ—Å—Ç–æ–º

- –∏–∑–≤–ª–µ–∫–∏—Ç–µ —Ç–µ–∫—Å—Ç –ø—É–±–ª–∏–∫–∞—Ü–∏–∏

- –∏–∑–≤–ª–µ–∫–∏—Ç–µ –ª—é–±—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤

4. –°—Ñ–æ—Ä–º–∏—Ä—É–π—Ç–µ —Ñ–∞–π–ª ***.csv**

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —à–∞–±–ª–æ–Ω:

````python
import vk_api
import csv
import pandas as pd
import matplotlib.pyplot as plt
````

````python


def scrape_wall_data(token, owner_id, count=10):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–∞—Ä—Å–∏–Ω–≥–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å–æ–æ–±—â–µ—Å—Ç–≤–∞ VK.

    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    - token: str, –≤–∞—à —Ç–æ–∫–µ–Ω –¥–æ—Å—Ç—É–ø–∞ VK API.
    - owner_id: int, ID —Å–æ–æ–±—â–µ—Å—Ç–≤–∞ (—Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, –Ω–∞–ø—Ä–∏–º–µ—Ä -123456).
    - count: int, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É–±–ª–∏–∫–∞—Ü–∏–π –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    - List[dict], —Å–ø–∏—Å–æ–∫ –ø—É–±–ª–∏–∫–∞—Ü–∏–π —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π: —Ç–µ–∫—Å—Ç, –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è, –ø—Ä–æ—Å–º–æ—Ç—Ä—ã, –ª–∞–π–∫–∏ –∏ —Ç.–¥.
    """
    try:
        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è VK API
        vk_session = vk_api.VkApi(token=token)
        vk = vk_session.get_api()

        # –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—É–±–ª–∏–∫–∞—Ü–∏–π —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ wall.get
        posts = vk.wall.get(owner_id=owner_id, count=count)['items']

        # –•—Ä–∞–Ω–∏–ª–∏—â–µ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö
        scraped_data = []

        # –ü–∞—Ä—Å–∏–Ω–≥
        for post in posts:
            post_data = {
                'post_id': post.get('id'),
                'text': post.get('text', '').replace('\n', ' '),  # –¢–µ–∫—Å—Ç –ø–æ—Å—Ç–∞
                'likes': post.get('likes', {}).get('count', 0),  # –õ–∞–π–∫–∏
                'views': post.get('views', {}).get('count', 0),  # –ü—Ä–æ—Å–º–æ—Ç—Ä—ã
                'comments': post.get('comments', {}).get('count', 0),  # –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏
                'reposts': post.get('reposts', {}).get('count', 0),  # –†–µ–ø–æ—Å—Ç—ã
                'images': [attachment['photo']['sizes'][-1]['url'] for attachment in post.get('attachments', [])
                           if attachment['type'] == 'photo']  # –°—Å—ã–ª–∫–∏ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
                ### –∑–¥–µ—Å—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é: –∏–ª–ª—é—Å—Ç—Ä–∞—Ü–∏–∏, —Ç–µ–∫—Å—Ç –ø—É–±–ª–∏–∫–∞—Ü–∏–∏, —Å—Ç–∞—Ç–∞... ###
            }
            scraped_data.append(post_data)

        return scraped_data

    except vk_api.exceptions.ApiError as e:
        print(f"Error accessing VK API: {e}")
        return []

# –î–æ–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è *.csv
def save_to_csv(data, file_name='vk_wall_posts.csv'):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤ —Ñ–æ—Ä–º–∞—Ç–µ CSV.

    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    - data: List[dict], –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è.
    - file_name: str, –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞.
    """

    with open(file_name, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['post_id', 'text', 'likes', 'views', 'comments', 'reposts', 'images'])
        writer.writeheader()
        for row in data:
            writer.writerow(row)

def stats(data):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–ø–∏—Å–∞–Ω–∏—è –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö.

    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    - data: List[dict], –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.
    """

    df = pd.DataFrame(data)

    # –í—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã
    print("–ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö:")
    print(df.head())

    plt.figure(figsize=(10, 6))
    plt.bar(df['post_id'], df['views'], label='Views', alpha=0.7, color='blue')
    plt.bar(df['post_id'], df['likes'], label='Likes', alpha=0.7, color='red')
    plt.xlabel('Post ID')
    plt.ylabel('Count')
    plt.title('Views and Likes per Post')
    plt.legend()
    plt.tight_layout()
    plt.show()



# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞—á–∏
def main():
    # –í–≤–µ–¥–∏—Ç–µ –≤–∞—à —Ç–æ–∫–µ–Ω VK API
    token = 'vk1.a.hVQEoaAzHhVUTvrrWWX6xk8c5XytIetzWaNsHHsK9GF15-B2_G_Fmh2RDyxlMsFR-inyaAn47-3h5coQWOPl00Szhvf1OTKqBZuoqxtPUR7T4D0w4WdQFBgH1AryIgdZXXnFouoIeGiABCby10J1m2iILtk2xGN7X3cvXh6Q1Scev4I6UnDQhxu3-T8ZXDXHp9ENVaYp_yHQ_5bJSoHvww'

    # –í–≤–µ–¥–∏—Ç–µ ID —Å–æ–æ–±—â–µ—Å—Ç–≤–∞
    owner_id = -299

    # –í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É–±–ª–∏–∫–∞—Ü–∏–π –¥–ª—è —Å–∫—Ä–µ–π–ø–∏–Ω–≥–∞
    num_posts = 10

    # –ó–∞–ø—É—Å—Ç–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–∫—Ä–µ–π–ø–∏–Ω–≥–∞
    wall_data = scrape_wall_data(token, owner_id, num_posts)

    if wall_data:
      # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –≤ CSV
      save_to_csv(wall_data, file_name='vk_wall_posts.csv')
      print("–î–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ vk_wall_posts.csv")

      stats(wall_data)
    else:
        print("No data")

if __name__ == "__main__":
    main()
````
